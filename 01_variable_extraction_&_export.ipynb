```python
# Function for extracting family-related variables
def family_variables():
  age_cols, sex_cols, rel_cols, mar_cols, grad_cols, race_cols, span_cols, \
  move_cols, movgrp_cols, famrel_cols, famtyp_cols, famline_cols \
  = [], [], [], [], [], [], [], [], [], [], [], []

  for i in range(1, 20):
    age_cols.append(f'AGE{i}')
    sex_cols.append(f'SEX{i}')
    rel_cols.append(f'REL{i}')
    mar_cols.append(f'MAR{i}')
    grad_cols.append(f'GRAD{i}')
    race_cols.append(f'RACE{i}')
    span_cols.append(f'SPAN{i}')
    move_cols.append(f'MOVE{i}')
    movgrp_cols.append(f'MOVERGRP{i}')
    famrel_cols.append(f'FAMREL{i}')
    famtyp_cols.append(f'FAMTYP{i}')
    famline_cols.append(f'FAMLINE{i}')

  fam_vars = age_cols + sex_cols + rel_cols + mar_cols + grad_cols + race_cols + span_cols + \
  move_cols + movgrp_cols + famrel_cols + famtyp_cols + famline_cols

  return fam_vars

fam_vars = family_variables()


# Function for extracting disabled-related variables
def disability_variables():
  pcare_cols, perrnd_cols, phear_cols, pmemry_cols, psee_cols, pwalk_cols \
  = [], [], [], [], [], []

  for i in range(1, 20):
    pcare_cols.append(f'PCARE{i}')
    perrnd_cols.append(f'PERRND{i}')
    phear_cols.append(f'PHEAR{i}')
    pmemry_cols.append(f'PMEMRY{i}')
    psee_cols.append(f'PSEE{i}')
    pwalk_cols.append(f'PWALK{i}')

  dis_vars = pcare_cols + perrnd_cols + phear_cols + pmemry_cols + psee_cols + pwalk_cols

  return dis_vars

dis_vars = disability_variables()


# Function for extracting income-related variables
def income_variables():
  intp_cols, oip_cols, pap_cols, retp_cols, semp_cols, ssip_cols, ssp_cols, wagp_cols \
  = [], [], [], [], [], [], [], []

  for i in range(1, 20):
    intp_cols.append(f'INTP{i}')
    oip_cols.append(f'OIP{i}')
    pap_cols.append(f'PAP{i}')
    retp_cols.append(f'RETP{i}')
    semp_cols.append(f'SEMP{i}')
    ssip_cols.append(f'SSIP{i}')
    ssp_cols.append(f'SSP{i}')
    wagp_cols.append(f'WAGP{i}')

  inc_vars = intp_cols + oip_cols + pap_cols + retp_cols + semp_cols + ssip_cols + ssp_cols + wagp_cols

  return inc_vars

inc_vars = income_variables()


# Function for extracting other variables 
def extract_variables(df0):
  extracted_vars = []
  for var in ['CONTROL', 'DIVISION', 'OMB13CBSA', 'WEIGHT',
              'INTSTATUS', 'TENURE',
              'HHCITSHP', 'HHNATVTY', 'HHRACE', 'HHSPAN', 'HHSEX', 'HHAGE', 'HHMAR', 'HHGRAD',
              'HHMOVE', 'SAMEHHLD',
              'HSHLDTYPE', 'MULTIGEN', 'NUMPEOPLE', 'NUMELDERS', 'NUMYNGKIDS', 'NUMOLDKIDS',
              'DISHH', 'NUMCARE', 'NUMERRND', 'NUMHEAR',
              'NUMMEMRY', 'NUMSEE', 'NUMWALK'] \
              + fam_vars + inc_vars + dis_vars \
              + ['BLD', 'GUTREHB', 'YRBUILT', 'UNITSIZE',
                 'TOTROOMS', 'BEDROOMS', 'KITCHENS', 'DINING',
                 'FINROOMS', 'LIVING', 'FAMROOMS', 'RECROOMS', 'DENS', 'OTHFN',
                 'BATHROOMS','LAUNDY', 'UFINROOMS',
                 'FINCP', 'HINCP','PERPOVLVL', 'POVLVLINC',
                 'TOTHCAMT', 'HOAAMT', 'INSURAMT', 'LOTAMT', 'MORTAMT', 'PROTAXAMT', 'RENT',
                 'UTILAMT', 'ELECAMT', 'GASAMT', 'OILAMT', 'OTHERAMT', 'TRASHAMT', 'WATERAMT',
                 'HUDSUB', 'RENTCNTRL', 'RENTSUB',
                 'MOVFORCE', 'MOVWHY',
                 'RMCHANGE', 'RMCOMMUTE', 'RMCOSTS', 'RMFAMILY', 'RMHOME', 'RMHOOD', 'RMJOB', 'RMOWNHH', 'RMOTHER',
                 'HRATE', 'NRATE',
                 'SEARCHAPRJ', 'SEARCHCRED ', 'SEARCHFAM', 'SEARCHGUAR',
                 'SEARCHLIST', 'SEARCHNEIGH', 'SEARCHNET', 'SEARCHOTH',
                 'SEARCHPUB', 'SEARCHREA', 'SEARCHSIGN', 'SEARCHSTOP',
                 'SEARCHTIME', 'SEARCHUNITS',
                 'MVG1COST', 'MVG1LOC', 'MVG1PER', 'MVG1STAT', 'MVG1TEN', 'MVG1TYPE',
                 'MVG2COST', 'MVG2LOC', 'MVG2PER', 'MVG2STAT', 'MVG2TEN', 'MVG2TYPE'
                 'MVG3COST', 'MVG3LOC', 'MVG3PER', 'MVG3STAT', 'MVG3TEN', 'MVG3TYPE',
                 'BATHEXCLU',
                 'HOTWATER',
                 'FUSEBLOW', 'NOWIRE', 'PLUGS',
                 'COLD', 'COLDCOST', 'COLDEQ', 'COLDEQFREQ', 'COLDHTCAP', 'COLDINSUL', 'COLDOTHER', 'COLDUTIL',
                 'NOTOIL', 'NOTOILFREQ', 'NOWAT', 'NOWATFREQ', 'SEWBREAK',
                 'ROACH', 'RODENT',
                 'FLOORHOLE', 'FNDCRUMB', 'PAINTPEEL', 'ROOFHOLE', 'ROOFSAG', 'ROOFSHIN',
                 'WALLCRACK', 'WALLSIDE', 'WALLSLOPE', 'WINBOARD', 'WINBROKE',
                 'LEAKI', 'LEAKIDK', 'LEAKIOTH', 'LEAKIPIPE', 'LEAKIPLUM', 'LEAKIWATH',
                 'LEAKO', 'LEAKOBASE ', 'LEAKOOTH', 'LEAKOROOF', 'LEAKOWALL',
                 'MOLDBASEM', 'MOLDBATH', 'MOLDBEDRM', 'MOLDKITCH', 'MOLDLROOM', 'MOLDOTHER',
                 'ADEQUACY', 'UPKEEP',
                 'DBLVEFORC', 'DBMISSMORT', 'DBEVICLK', 'DBEVICNOTE', 'DBEVICTHT', 'DBMISSRENT', 'DBUTBILL',
                 'NHQSCHOOL', 'NHQPCRIME', 'NHQSCRIME', 'NHQPUBTRN', 'NHQRISK', 'RATINGNH', 'RATINGHS']:

    if var in df0.columns:
      extracted_vars.append(var)

  extracted_df = df0[extracted_vars].copy()

  return extracted_df
```

```python
# Function for data manipulation
def calculate(df0):
  df = df0.copy()

  # Convert character to numeric
  tmp_var_list1 = ['DIVISION', 'OMB13CBSA', 'TENURE',
                   'HHRACE', 'HHSPAN', 'HHSEX', 'HHCITSHP', 'HHGRAD', 'HHMAR',
                   'MULTIGEN', 'BATHROOMS', 'BLD', 'RENTCNTRL', 'HUDSUB']

  for var in tmp_var_list1:
    if df[var].dtype not in ['float64', 'int64']:
      df[var] = pd.to_numeric(df[var].str.replace("'", ""))

  # Grossrent
  rent_categories = ['RENT', 'ELECAMT', 'GASAMT', 'OILAMT', 'OTHERAMT', 'TRASHAMT', 'WATERAMT']
  for var in rent_categories:
    df.loc[df[var] < 0, var] = np.nan

  df = df.assign(
      GROSSRENT = np.where(df['RENT'].isnull(), np.nan, np.nansum(df[rent_categories], axis=1))
  )

  # Incomes
  inc_categories = []
  for var in inc_vars:
    if var in df.columns:
      if df[var].dtype not in ['float64', 'int64']:
        df[var] = pd.to_numeric(df[var].str.replace(".", ""))
      inc_categories.append(var)

  incs = np.where(df[inc_categories] == -9, np.nan,
         np.where(df[inc_categories] == -6, 0, df[inc_categories]))

  incs_df = pd.DataFrame(incs, columns=inc_categories).abs()

  df = df.assign(
      INCOME = np.where(incs_df.isnull().all(axis=1), np.nan, np.nansum(incs_df, axis=1))
  )

  df = df.assign(
      LN_INCOME = np.log1p(df['INCOME']),
      INCOME_CATEGORY = np.where((df['PERPOVLVL'] > 0) & (df['PERPOVLVL'] <= 100), 1,
                        np.where((df['PERPOVLVL'] > 100) & (df['PERPOVLVL'] <= 200), 2,
                        np.where((df['PERPOVLVL'] > 200) & (df['PERPOVLVL'] <= 400), 3,
                        np.where(df['PERPOVLVL'] > 400, 4, np.nan))))
  )

  # RIR (Rent-to-Income Ratio)
  df = df.assign(
      RIR = np.where((df['INCOME'] != 0) & (~df['INCOME'].isnull()) & (~df['GROSSRENT'].isnull()), df['GROSSRENT'] / (df['INCOME'] / 12) * 100,
            np.where((df['INCOME'] == 0) & (~df['GROSSRENT'].isnull()), df['GROSSRENT'] / 1 * 100, np.nan))
  )

  # Household members
  df = df.assign(
      NUM_SENIORS = np.where(df['NUMELDERS'] >= 0, df['NUMELDERS'], np.nan),
      NUM_YNGKIDS = np.where(df['NUMYNGKIDS'] >= 0, df['NUMYNGKIDS'], np.nan),
      NUM_OLDKIDS = np.where(df['NUMOLDKIDS'] >= 0, df['NUMOLDKIDS'], np.nan),
      NUM_MEMBERS = np.where(df['NUMPEOPLE'] >= 0, df['NUMPEOPLE'], np.nan)
  )

  df = df.assign(
      SENIORS = (df['NUMELDERS'] >= 1).astype(int),
      YNGKIDS = (df['NUMYNGKIDS'] >= 1).astype(int),
      OLDKIDS = (df['NUMOLDKIDS'] >= 1).astype(int)
  )

  # Metropolitan area
  df = df.assign(
      METRO = (df['OMB13CBSA'] != 99999).astype(int)
  )

  # Division
  df = df.assign(
      DIV4 = np.where(df['DIVISION'].isin([1, 2]), 1,
             np.where(df['DIVISION'].isin([3, 4]), 2,
             np.where(df['DIVISION'].isin([5, 6, 7]), 3,
             np.where(df['DIVISION'].isin([8, 9]), 4, np.nan))))
  )

  # Householder race
  df = df.assign(
      RACE_CATEGORY = np.where(df['HHRACE'] == 1, 1, # White
                      np.where(df['HHRACE'] == 2, 2, # Black
                      np.where(df['HHRACE'] == 4, 3, # Asian
                      np.where((df['HHRACE'] == 3) | (df['HHRACE'] >= 5), 4, np.nan)))), # Other races
      LATIN = np.where(df['HHSPAN'] == 1, 1, np.where(df['HHSPAN'] == 2, 0, np.nan))
  )

  # Householder native born
  df = df.assign(
      NON_US_CITIZEN = np.where(df['HHCITSHP'] >= 5, 1,
                       np.where((df['HHCITSHP'] >= 1) & (df['HHCITSHP'] <= 4), 0, np.nan))
  )

  # Householder age
  df = df.assign(
      HH_AGE = np.where(df['HHAGE'] > 0, df['HHAGE'], np.nan),
  )

  # Householder sex
  df = df.assign(
      FEMALE = (df['HHSEX'] == 2).astype(int),
      MALE = (df['HHSEX'] == 1).astype(int)
  )

  # Householder high school & college degree
  df = df.assign(
      GRAD_CATEGORY = np.where((df['HHGRAD'] >= 1) & (df['HHGRAD'] <= 38), 1, # Under_highschool
                      np.where(df['HHGRAD'] == 39, 2, # Highschool
                      np.where((df['HHGRAD'] >= 40) & (df['HHGRAD'] <= 43), 3, # College
                      np.where(df['HHGRAD'] == 44, 4, # Bachelor
                      np.where((df['HHGRAD'] >= 45) & (df['HHGRAD'] <= 47), 5, np.nan))))) # Master_or_higher
  )

  # Marital status of Householder
  df = df.assign(
      MAR_CATEGORY = np.where(df['HHMAR'].isin([1, 2]), 1, # Married
                     np.where(df['HHMAR'].isin([3, 4, 5]), 2, # Separated
                     np.where(df['HHMAR'].isin([6]), 3, np.nan))) # Never_married
  )

  # Type of multigenerational household
  df = df.assign(
      MULTI_GEN = ((df['MULTIGEN'] >= 2) & (df['MULTIGEN'] <= 6)).astype(int)
  )

  # Disabled
  dis_vars_lists = {}
  dis_vars_new = {}
  dis_p = []

  for i in range(1, 20):
    dis_vars_lists[i] = [f'PCARE{i}', f'PERRND{i}', f'PHEAR{i}', f'PMEMRY{i}', f'PSEE{i}', f'PWALK{i}']

    if set(dis_vars_lists[i]).issubset(set(df.columns)):
      for var in dis_vars_lists[i]:
        if df[var].dtype not in ['float64', 'int64']:
          df[var] = pd.to_numeric(df[var].str.replace("'", "").str.replace(" ", ""))

      dis_vars_new[i] = dis_vars_lists[i]

      dis_p.append(np.where((df[dis_vars_new[i]] == 1).any(axis=1), 1, 0))

  df = df.assign(
      NUM_DISABLED = np.sum(dis_p, axis=0)
  )

  df = df.assign(
      DISABLED = (df['NUM_DISABLED'] >= 1).astype(int)
  )

  # Stay period
  df = df.assign(
      STAY_PERIOD = 2021 - df['HHMOVE']
  )

  df = df.assign(
      STAY_CATEGORY = np.where(df['STAY_PERIOD'] >= 10, 3,
                      np.where(df['STAY_PERIOD'] >= 5, 2,
                      np.where(df['STAY_PERIOD'] >= 0, 1, np.nan)))
  )

  # Type of housing unit
  df = df.assign(
      UNIT_CATEGORY = np.where((df['BLD'] >= 2) & (df['BLD'] <= 3), 1,
                      np.where((df['BLD'] >= 4) & (df['BLD'] <= 9), 2,
                      np.where((df['BLD'] == 1) | (df['BLD'] == 10), 3, np.nan)))
  )

  # House age
  df = df.assign(
      HOUSE_AGE = 2021 - df['YRBUILT']
  )

  # Rent conrtol & Renter Subsidy
  df = df.assign(
      RENT_CONTROL = (df['RENTCNTRL'] == 1).astype(int),
      HUD_CATEGORY = np.where(df['HUDSUB'] == 1, 1,
                     np.where(df['HUDSUB'] == 2, 2,
                     np.where(df['HUDSUB'] == 3, 3, np.nan)))
  )

  # Neighborhood
  for var in ['NHQSCHOOL', 'NHQPCRIME', 'NHQSCRIME', 'NHQPUBTRN', 'NHQRISK', 'RATINGNH', 'RATINGHS']:
    if df[var].dtype not in ['float64', 'int64']:
      df[var] = pd.to_numeric(df[var].str.replace("'", "").str.replace("N", "").str.replace("M", "").str.replace(" ", ""))

  df = df.assign(
      NOGOODSCHOOL = (df['NHQSCHOOL'] == 2).astype(int),
      NOGOODPTRNS = (df['NHQPUBTRN'] == 2).astype(int),
      HIGH_PCRIME = (df['NHQPCRIME'] == 1).astype(int),
      HIGH_SCRIME = (df['NHQSCRIME'] == 1).astype(int),
      HIGH_RISK = (df['NHQRISK'] == 1).astype(int)
  )

  # Housing Adequacy
  for var in ['ADEQUACY', 'UPKEEP', 'HOTWATER', 'BATHEXCLU', 'COLD', 'COLDEQFREQ', 'NOWIRE', 'PLUGS', 'FUSEBLOW']:
    if df[var].dtype not in ['float64', 'int64']:
      df[var] = pd.to_numeric(df[var].str.replace("'", ""))

  # Housing insecurity
  hi_vars = ['DBLVEFORC', 'DBMISSMORT', 'DBEVICLK', 'DBEVICNOTE', 'DBEVICTHT', 'DBMISSRENT', 'DBUTBILL']

  if set(hi_vars).issubset(set(df.columns)):
    for var in hi_vars:
      if df[var].dtype not in ['float64', 'int64']:
        df[var] = pd.to_numeric(df[var].str.replace("'", ""))

  # Composit index of housing insecurity
  df = df.assign(
      HI_EVICT_RISK1 = (df['DBEVICLK'] == 1).astype(int),   # likelihood of leaving home within the next 2 months because of eviction - 1.Very likely 2.Somewhat likely 3.Not very likely
      HI_EVICT_THREAT = (df['DBEVICTHT'] == 1).astype(int),   # Have been threatened with eviction in the last 3 months - 1.Yes 2.No
      HI_RENT13 = ((df['DBMISSRENT'] >= 1) & (df['DBMISSRENT'] <= 3)).astype(int),   # Unable to pay 1-3 of the last three months
      HI_UTIL1 = (df['DBUTBILL'] == 1).astype(int),   # Received notice, utilities shut off (1)
      HI_ADEQ_SEV = (df['ADEQUACY'] == 3).astype(int),
      HI_UPK3 = ((df['UPKEEP'] >= 2) & (df['UPKEEP'] <= 3)).astype(int),   # 3 or more upkeep problems
      HI_BADHOME4 = ((df['RATINGHS'] >= 1) & (df['RATINGHS'] <= 4)).astype(int),
      HI_RIR50 = (df['RIR'] > 50).astype(int),
      HI_POVERTY133 = ((df['RIR'] > 30) & (df['PERPOVLVL'] > 0) & (df['PERPOVLVL'] < 133)).astype(int),
  )

  df = df.assign(
      COMPOSITE_HI = df[['HI_EVICT_RISK1', 'HI_EVICT_THREAT', 'HI_RENT13', 'HI_UTIL1', 'HI_ADEQ_SEV', 'HI_UPK3', 'HI_BADHOME4', 'HI_RIR50', 'HI_POVERTY133']].sum(axis=1),
      CHI_RI = ((df['HI_EVICT_RISK1'] == 1) | (df['HI_EVICT_THREAT'] == 1) | (df['HI_RENT13'] == 1) | (df['HI_UTIL1'] == 1)).astype(int),
      CHI_UI = ((df['HI_ADEQ_SEV'] == 1) | (df['HI_UPK3'] == 1) | (df['HI_BADHOME4'] == 1)).astype(int),
      CHI_UA = ((df['HI_RIR50'] == 1) | (df['HI_POVERTY133'] == 1)).astype(int),
  )

  conditions = [
      (df['CHI_RI'] == 1) & (df['CHI_UI'] == 0) & (df['CHI_UA'] == 0),
      (df['CHI_RI'] == 0) & (df['CHI_UI'] == 1) & (df['CHI_UA'] == 0),
      (df['CHI_RI'] == 0) & (df['CHI_UI'] == 0) & (df['CHI_UA'] == 1),
      (df['CHI_RI'] == 1) & (df['CHI_UI'] == 1) & (df['CHI_UA'] == 0),
      (df['CHI_RI'] == 1) & (df['CHI_UI'] == 0) & (df['CHI_UA'] == 1),
      (df['CHI_RI'] == 0) & (df['CHI_UI'] == 1) & (df['CHI_UA'] == 1),
      (df['CHI_RI'] == 1) & (df['CHI_UI'] == 1) & (df['CHI_UA'] == 1)
  ]

  choices = ['RI', 'IU', 'UA', 'RI + UI', 'RI + UA', 'UA + UA', 'RI + UA + UA']

  df = df.assign(
      COMPOSITE_HI2 = np.where(df['COMPOSITE_HI'] >= 3, 3, df['COMPOSITE_HI']),
      CHI_CATEGORY = np.select(conditions, choices, pd.NA)
  )

  return df
```

```python
# Function for converting to int type
def convert_int(df0):
  df = df0.copy()

  df['COMPOSITE_HI2'] = df['COMPOSITE_HI2'].astype(int)
  df['NUM_MEMBERS'] = df['NUM_MEMBERS'].astype(int)
  df[['STAY_PERIOD', 'HOUSE_AGE']] = df[['STAY_PERIOD', 'HOUSE_AGE']].astype(float)
  df[['DIVISION', 'DIV4']] = df[['DIVISION', 'DIV4']].astype(int)
  df[['LATIN', 'NON_US_CITIZEN']] = df[['LATIN', 'NON_US_CITIZEN']].astype(int)
  df['RACE_CATEGORY'] = df['RACE_CATEGORY'].astype(int)
  df['GRAD_CATEGORY'] = df['GRAD_CATEGORY'].astype(int)
  df['MAR_CATEGORY'] = df['MAR_CATEGORY'].astype(int)
  df['STAY_CATEGORY'] = df['STAY_CATEGORY'].astype(int)
  df['RENT_CONTROL'] = df['RENT_CONTROL'].astype(int)
  df['HUD_CATEGORY'] = df['HUD_CATEGORY'].astype(int)
  df['UNIT_CATEGORY'] = df['UNIT_CATEGORY'].astype(int)

  return df


# Function for one-hot encoding 1
def ohe_manual(df):
  df = df.assign(
      WHITE = (df['RACE_CATEGORY'] == 1).astype(int),
      BLACK = (df['RACE_CATEGORY'] == 2).astype(int),
      ASIAN = (df['RACE_CATEGORY'] == 3).astype(int),
      OTHER_RACE = (df['RACE_CATEGORY'] == 4).astype(int)
  )

  df = df.assign(
      MARRIED = (df['MAR_CATEGORY'] == 1).astype(int),
      SEPARATED = (df['MAR_CATEGORY'] == 2).astype(int),
      NEVER_MARRIED = (df['MAR_CATEGORY'] == 3).astype(int)
  )

  df = df.assign(
      HUD_HOU = (df['HUD_CATEGORY'] == 1).astype(int),
      HUD_HCV = (df['HUD_CATEGORY'] == 2).astype(int),
      HUD_NON = (df['HUD_CATEGORY'] == 3).astype(int)
  )

  df = df.assign(
      ONE_FAMILY_HOUSE = (df['UNIT_CATEGORY'] == 1).astype(int),
      APARTMENT = (df['UNIT_CATEGORY'] == 2).astype(int),
      MOBILE_HOUSE = (df['UNIT_CATEGORY'] == 3).astype(int)
  )

  df = df.assign(
      DIV4_1 = (df['DIV4'] == 1).astype(int),
      DIV4_2 = (df['DIV4'] == 2).astype(int),
      DIV4_3 = (df['DIV4'] == 3).astype(int),
      DIV4_4 = (df['DIV4'] == 4).astype(int)
  )

  return df


# Function for one-hot encoding 2 (education level)
def ohe_manual_edu(df):
  df = df.assign(
      UNDSCHOOL = (df['GRAD_CATEGORY'] == 1).astype(int),
      HIGHSCHOOL = (df['GRAD_CATEGORY'] == 2).astype(int),
      COLLEGE = (df['GRAD_CATEGORY'] == 3).astype(int),
      BACHELOR = (df['GRAD_CATEGORY'] == 4).astype(int),
      MASTER = (df['GRAD_CATEGORY'] == 5).astype(int)
  )

  df = df.assign(
      STAY_0_5 = (df['STAY_CATEGORY'] == 1).astype(int),
      STAY_5_10 = (df['STAY_CATEGORY'] == 2).astype(int),
      STAY_10_MORE = (df['STAY_CATEGORY'] == 3).astype(int)
  )

  return df
```
